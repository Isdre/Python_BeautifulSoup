```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
r = requests.get("https://olympics.com/en/olympic-games/tokyo-2020/medals")
soup = bs(r.content)
table = soup.find("div",class_="Tablestyles__CommonGrid-xjyvs9-1 Tablestyles__Content-xjyvs9-3 kSOeZM dzUKLb")
data = {"Flag":[],"Shortcut":[],"Country":[],"Gold":[],"Silver":[],"Bronze":[],"Total":[]}
images = table.find_all("div",class_="styles__Flag-sc-fehzzg-2 gmCcuH")
countries = table.find_all("div",class_="styles__CountryWrapper-sc-fehzzg-4 dVfDKJ")
gold_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Gold")
silver_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Silver")
bronze_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Bronze")
total_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="")
```

```python
for i,c,g,s,b,t in zip(images,countries,gold_medals,silver_medals,bronze_medals,total_medals):
    flags = i.find("source")
    flag = flags["data-srcset"].split(" ")[1]
    data["Flag"].append(flag)
    short = c.text.split(" ")
    data["Shortcut"].append(short[0])
    full_name = " ".join(short[1:])
    data["Country"].append(full_name)
    data["Gold"].append(int(g.text.replace("-","0")))
    data["Silver"].append(int(s.text.replace("-","0")))
    data["Bronze"].append(int(b.text.replace("-","0")))
    data["Total"].append(int(t.text.replace("-","0")))
```

Getting data for all olimpics

```python
def get_data_from_olimpic_web(url):
    try:
        r = requests.get(url)
        soup = bs(r.content)
        table = soup.find("div",class_="Tablestyles__CommonGrid-xjyvs9-1 Tablestyles__Content-xjyvs9-3 kSOeZM dzUKLb")
        data = {"Flag":[],"Shortcut":[],"Country":[],"Gold":[],"Silver":[],"Bronze":[],"Total":[]}
        images = table.find_all("div",class_="styles__Flag-sc-fehzzg-2 gmCcuH")
        countries = table.find_all("div",class_="styles__CountryWrapper-sc-fehzzg-4 dVfDKJ")
        gold_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Gold")
        silver_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Silver")
        bronze_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="Bronze")
        total_medals = table.find_all("div",class_="Medalstyles__Wrapper-sc-1tu6huk-0 kXFxTL",title="")
        for i,c,g,s,b,t in zip(images,countries,gold_medals,silver_medals,bronze_medals,total_medals):
            flags = i.find("source")
            flag = flags["data-srcset"].split(" ")[1]
            data["Flag"].append(flag)
            short = c.text.split(" ")
            data["Shortcut"].append(short[0])
            full_name = " ".join(short[1:])
            data["Country"].append(full_name)
            data["Gold"].append(int(g.text.replace("-","0")))
            data["Silver"].append(int(s.text.replace("-","0")))
            data["Bronze"].append(int(b.text.replace("-","0")))
            data["Total"].append(int(t.text.replace("-","0")))
        return data
    except Exception as e:
        return e
```

```python
data_sample = {"Flag":[],"Shortcut":[],"Country":[],"Gold":[],"Silver":[],"Bronze":[],"Total":[],"Place":[],"Year":[]}
olimpics_table = pd.DataFrame(data=data_sample)
hollow_url = "https://olympics.com/en/olympic-games/{}/medals"
r = requests.get("https://olympics.com/en/olympic-games/olympic-results")
soup = bs(r.content)
table = soup.find("section",class_="Gridstyles__GridContainer-sc-1p7u4tu-0 fLQzGx styles__TabContentNotPaginated-sc-z55d96-6 ezZekU")
olimpics = table.find_all("section",class_="styles__ItemLink-sc-111k07d-2 ipqQzr link-item")
for olimpic in olimpics:
    paragraph = olimpic.find("p")
    url = paragraph.text.replace(" ","-").lower()
    try:
        data = get_data_from_olimpic_web(hollow_url.format(url))
        tab = pd.DataFrame(data=data)
        print(url)
        place_and_year = url.split("-")
        tab["Place"] = " ".join(place_and_year[:-1])
        print(" ".join(place_and_year[:-1]))
        tab["Year"] = place_and_year[-1]
        print(place_and_year[-1])
        olimpics_table = pd.concat([olimpics_table,tab])
    except Exception as e:
        print(e)
```

```python
olimpics_table.reset_index(inplace=True)
olimpics_table.to_csv("olimpics_table.csv")
print(olimpics_table)
```
