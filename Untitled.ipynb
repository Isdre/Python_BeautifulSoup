```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
r = requests.get("https://en.wikipedia.org/wiki/2020_Summer_Olympics_medal_table")
soup = bs(r.content)
content = soup.find(class_="wikitable sortable plainrowheaders jquery-tablesorter")
rows = content.find_all("tr")
for row in rows:
    print(row.prettify())
```

```python
import pandas as pd
data = {}
last = 0
for index,row in enumerate(rows):
    if index == 0:
        heads = row.find_all("th")
        for head in heads:
            data[head.get_text()]=[]
    else:
        r = index
        title = row.find("th")
        stats = row.find_all("td")
#         print(title.get_text())
        data['NOC'].append(title.get_text())
        z = 1
        if len(stats) == 4:
            r = last
            z=0
        data['Rank'].append(r)
        data['Gold'].append(stats[z].get_text())
        data['Silver'].append(stats[z+1].get_text())
        data['Bronze'].append(stats[z+2].get_text())
        data['Total'].append(stats[z+3].get_text())
        last = r
    
tab = pd.DataFrame(data=data)
tab.drop([93],inplace=True)
tab.rename(columns={"NOC":"Country"},inplace=True)
tab
```
